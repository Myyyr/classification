```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def file_to_np(path):
    f = open(path, "r")
    txt = f.read()
    f.close()
    txt = txt.split("\n")
    dt = np.array(txt[1:-1]).astype(np.float)
    return dt
    
```

### ResNet-18

```python
dt = file_to_np("results/fold_resnet18-MSSch/train_epoch_scores.csv")
plt.plot(dt)
dt = file_to_np("results/fold_resnet18-MSSch/dev_epoch_scores.csv")
plt.plot(dt)

plt.show()
```
